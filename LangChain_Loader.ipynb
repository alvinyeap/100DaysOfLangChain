# LLMChain
Document loaders are the first step in getting new knowledge in to an LLM.

Here we're using LangChain's GutenbergLoader to load a free book Rene Descarte's Discourse on Method where he first says:
- "I think, therefore I am"
- "It is not enough to have a good mind. The main thing is to use it well."

Such wise words!

```python
# Install LangChain
!pip install langchain
```

```python
from langchain.document_loaders import GutenbergLoader

# Create the loader.
loader = GutenbergLoader('https://www.gutenberg.org/cache/epub/59/pg59.txt')
# And load the book! Too easy!
data = loader.load()
```

```python
# We get back a list of Documents.
# We only have one document.
# Let's see it's metadata
print(data[0].metadata)
```

```python
# And let's check out it's contents
print(data[0].page_content)
```
